# Visualization with Hierarchical Clustering and t-SNE

## Hierarchical clustering

```python
import pandas as pd
import numpy as np
eurovision = pd.read_csv('datasets/eurovision-2016.csv')
eurovision.head()
```

## Pre-process eurovision data

```python
eurovision.fillna(0, inplace=True)
eurovision.head()
```

```python
eurovisionPivot = pd.pivot_table(eurovision, values='Jury Points', index='From country', columns='To country', aggfunc=np.sum).fillna(0)
eurovisionPivot
```

```python
country_names = eurovisionPivot.index.values.astype(str)
country_names
```

```python
samples = eurovisionPivot.values.astype(int)
samples.shape
```

## Hierarchical clustering: complete and single linkage

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Calculate the linkage: mergings
mergings = linkage(samples, method='complete')
mergings
```

```python
# Plot the dendrogram, using country_names as labels
dendrogram(mergings,
           labels=country_names,
           leaf_rotation=90,
           leaf_font_size=6,
)
plt.show()
```

```python
# Calculate the linkage: mergings
mergings = linkage(samples, method='single')

# Plot the dendrogram
dendrogram(mergings, labels=country_names, leaf_rotation=90, leaf_font_size=6)
plt.show()
```

## with seeds data

```python
seeds = pd.read_csv('datasets/Grains/seeds.csv', header=None).values[:, 0:7]
seeds
```

```python
wheats = np.array(['Kama', 'Rosa', 'Canadian'])
targets = pd.read_csv('datasets/Grains/seeds.csv', header=None).values[:, 7].astype(int) - 1
varieties = wheats[targets]
varieties
```

```python
variety_numbers = pd.read_csv('datasets/Grains/seeds.csv', header=None).values[:, 7].astype(int)
variety_numbers
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(seeds, varieties, test_size = 0.75, random_state=42, stratify=varieties)
```

```python
# Calculate the linkage: mergings
mergings = linkage(X_train, method='complete')

# Plot the dendrogram, using varieties as labels
dendrogram(mergings,
           labels=y_train,
           leaf_rotation=90,
           leaf_font_size=6,
)
plt.show()
```

## t-SNE visualization of grain dataset

```python
'''
t-SNE
- t-SNE stans for t-distributed stochastic neighbor embedding

Cons of t-SNE
1. t-SNE does not have separate fit and transform methods. It means that you can't extend t-SNE map to include new smaples.
2. Learning rate
- Choose learning rate for the dataset
- Wrong choice: points bunch together
- Try values between 50 and 200
3. Different every time
- t-SNE features are different every time
- e.g. t-SNE plots of the same samples generate different plots when using the same code.
'''


# Import TSNE
from sklearn.manifold import TSNE
from sklearn import datasets

iris = datasets.load_iris()

# Create a TSNE instance: model
model = TSNE(learning_rate=200)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(iris.data)

# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1st feature: ys
ys = tsne_features[:,1]

# Scatter plot, coloring by variety_numbers
plt.scatter(xs, ys, c=iris.target)
plt.show()
```

```python
# Create a TSNE instance: model
model = TSNE(learning_rate=200)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(seeds)

# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1st feature: ys
ys = tsne_features[:,1]

# Scatter plot, coloring by variety_numbers
plt.scatter(xs, ys, c=variety_numbers)
plt.show()
```

## A t-SNE map of the stock market

```python
movements = pd.read_csv('datasets/company-stock-movements-2010-2015-incl.csv').values[:, 1:964].astype(float)
movements.shape
```

```python
companies = pd.read_csv('datasets/company-stock-movements-2010-2015-incl.csv').values[:, 0]
companies.shape
```

```python
# Import normalize
from sklearn.preprocessing import normalize

# Normalize the movements: normalized_movements
normalized_movements = normalize(movements)

# Calculate the linkage: mergings
mergings = linkage(normalized_movements, method='complete')

# Plot the dendrogram
dendrogram(mergings,
           labels=companies,
           leaf_rotation=90,
           leaf_font_size=6
)
plt.show()
```

```python
# Import TSNE
#from sklearn.manifold import TSNE

# Create a TSNE instance: model
model = TSNE(learning_rate=50)

# Apply fit_transform to normalized_movements: tsne_features
tsne_features = model.fit_transform(normalized_movements)

# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1th feature: ys
ys = tsne_features[:,1]

# Scatter plot
plt.scatter(xs, ys, alpha=0.5)

# Annotate the points
for x, y, company in zip(xs, ys, companies):
    plt.annotate(company, (x, y), fontsize=5, alpha=0.75)
plt.show()
```
